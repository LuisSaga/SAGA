```python
import pandas as pd
import numpy as np
import requests

# Ruta al archivo Excel
archivo_excel = r'E:\Inteligencia Comercial\Tablas Catalogo\Ubicacion_Estaciones.xlsx'

# Cargar datos desde el archivo Excel a un DataFrame de pandas
try:
    # Lee el archivo Excel y carga la hoja 'base_global' en un DataFrame
    Estaciones = pd.read_excel(archivo_excel, sheet_name='Ubicacion')
    
except Exception as e:
    print("Error al cargar los datos desde el archivo Excel:", e)
```

```python
import folium
from folium import CircleMarker, Popup
from IPython.display import display

# Coordenadas del centro
centro_latitud, centro_longitud = 25.675456504772935, -100.32213321988202

# Filtrar filas con valores numéricos en 'Latitud' y 'Longitud'
Estaciones = Estaciones.dropna(subset=['Latitud', 'Longitud'])

# Inicializa el mapa centrado en la ubicación especificada
mapa = folium.Map(location=[centro_latitud, centro_longitud], zoom_start=9)

# Agrega círculos y marcadores al mapa usando el DataFrame
for index, row in Estaciones.iterrows():
    etiqueta = f"{row['Estaciones']} - {row['Unidad']}"  # Puedes personalizar la etiqueta según tus columnas
    
    # Crea un círculo con centro en la ubicación de la estación y radio de 100 metros
    circle = CircleMarker(location=[row['Latitud'], row['Longitud']], radius=5, color='#E1251B', fill=True, fill_color='#E1251B')
    
    # Agrega una etiqueta emergente (popup) al círculo
    popup = Popup(etiqueta, parse_html=True)
    popup.add_to(circle)
    
    # Agrega el círculo al mapa
    circle.add_to(mapa)

    # Agrega un marcador con el mismo color que el círculo
    marker = folium.Marker(location=[row['Latitud'], row['Longitud']], icon=folium.Icon(color='red'), tooltip=etiqueta)
    marker.add_to(mapa)

# Muestra el mapa en el entorno de IPython
display(mapa)


```

```python
# Guarda el mapa como un archivo HTML
mapa.save("mapa_exportado.html")
```
